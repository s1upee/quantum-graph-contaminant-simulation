```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Step 1: Create a graph
G = nx.erdos_renyi_graph(n=10, p=0.3)  # 10 nodes, 30% chance of edge between any two

# Step 2: Pick a random node to start contamination
infected = set()
start_node = random.choice(list(G.nodes))
infected.add(start_node)

# Step 3: Simulate one round of spreading (30% chance to infect neighbors)
for neighbor in G.neighbors(start_node):
    if random.random() < 0.3:  # 30% chance
        infected.add(neighbor)

# Step 4: Draw the graph and highlight infected nodes
colors = ['red' if node in infected else 'lightblue' for node in G.nodes]

plt.figure(figsize=(6,6))
nx.draw(G, with_labels=True, node_color=colors, node_size=600)
plt.title("Contaminant Spread: Red = Infected")
plt.show()

```
